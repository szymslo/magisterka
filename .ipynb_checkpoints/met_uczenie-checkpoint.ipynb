```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sklearn
from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv('./data/ktw_modified.csv') #wczytanie danych
```

```python
data.set_index(keys=['Rok','Miesiac','Dzien'], inplace = True) #ustawienie multiindeksu na rok-miesiac-dzien
```

```python
data.sort_index(inplace = True) #sortowanie po indeksie
```

```python
data['Typ_opadu'] = data['Typ_opadu'].round(0).astype(int) #zaokrąglenie i rzutowanie na int
```

```python
data #wyswietlenie gotowych danych
```

```python
d1 = {'Temp': data.Tmax}
d2 = {'Temp': data.Tmin}
tmax = pd.DataFrame(d1)
tmin = pd.DataFrame(d2)
```

```python
fullt = tmax.append(tmin, ignore_index=False)
```

```python
fullt
```

```python
model=IsolationForest(n_estimators=100, max_samples='auto', contamination=float(0.002))
```

```python
model.fit(fullt[['Temp']])
```

```python
fullt['Score']=model.decision_function(fullt[['Temp']])
fullt['IsOutlier']=model.predict(fullt[['Temp']])
```

```python
# .fit i .predict czyli trenowanie i testowanie jest na tych samych danych
```

```python
# zwracamy wyliczony przez funkcje decyzyjną algorytmu wynik "Score" dla każdego punktu oraz decyzję "IsOutlier"
# -1 znaczy ze jest outlierem
```
