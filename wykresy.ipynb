```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('./data/ktw_modified.csv') #wczytanie danych
```

```python
data.set_index(keys=['Rok','Miesiac','Dzien'], inplace = True) #ustawienie multiindeksu na rok-miesiac-dzien
```

```python
data.sort_index(inplace = True) #sortowanie po indeksie
```

```python
data #wyswietlenie gotowych danych
```

```python
tmax = pd.Series(data.reset_index().Tmax)
tmin = pd.Series(data.reset_index().Tmin)
```

```python
fullt = tmax.append(tmin, ignore_index=True)
```

```python
color = dict(boxes='DarkBlue', whiskers='Black', medians='DarkGreen', caps='Gold')
boxprops = dict(linestyle='-', linewidth=2)
whiskerprops = dict(linestyle='-', linewidth=1.75)
medianprops = dict(linestyle='-', linewidth=2.5)
capprops = dict(linestyle='-', linewidth=1.5)
```

```python
pudelkowyFull = fullt.plot.box(figsize=(7,9),label='2001-2019', color=color, boxprops=boxprops, 
                               medianprops=medianprops, whiskerprops=whiskerprops, capprops=capprops)
pudelkowyFull.set_title("Katowice, temperatura powietrza w latach 2001-2019", weight='bold', size=10)
pudelkowyFull.set_ylabel("Temperatura [Â°C]", labelpad=10, weight='bold', size=9)
pudelkowyFull.spines['right'].set_visible(False)
pudelkowyFull.spines['top'].set_visible(False)
pudelkowyFull.grid(axis='y', alpha = 0.2)
```

```python
histFull = fullt.plot.hist(figsize=(17,10), bins=fullt.nunique(), alpha=0.9, rwidth=0.85, color='mediumseagreen')
histFull.set_title("Katowice, temperatura powietrza w latach 2001-2019", weight='bold', size=14)
histFull.set_ylabel("CzÄ™stotliwoÅ›Ä‡", labelpad=10, weight='bold', size=12)
histFull.set_xlabel("Temperatura [Â°C]", labelpad=10, weight='bold', size=12)
histFull.tick_params(axis='both', which="both", bottom=False, top=False, right=False)
histFull.spines['right'].set_visible(False)
histFull.spines['top'].set_visible(False)
histFull.grid(axis='y', alpha = 0.2)
```

```python
opad = pd.DataFrame(data[['Opad','Typ_opadu']])
```

```python
opad.rename(columns={'Opad':'Suma','Typ_opadu':'Typ'}, inplace=True)
```

```python
existingOpad = opad.loc[opad.Suma > 1 ,'Suma']
```

```python
pudelkowyOpad = existingOpad.plot.box(figsize=(7,11),label='2001-2019', whis=3, color=color, boxprops=boxprops, 
                               medianprops=medianprops, whiskerprops=whiskerprops, capprops=capprops)
pudelkowyOpad.set_title("Dobowa suma opadu w Katowicach w latach 2001-2019", weight='bold', size=10)
pudelkowyOpad.set_ylabel("Opad [mm]", labelpad=10, weight='bold', size=9)
pudelkowyOpad.spines['right'].set_visible(False)
pudelkowyOpad.spines['top'].set_visible(False)
pudelkowyOpad.grid(axis='y', alpha = 0.2)
```
